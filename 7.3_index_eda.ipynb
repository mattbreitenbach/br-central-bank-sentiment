```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")
```

```python

df_optimism = pd.read_excel("./data/processed/index_optimism.xlsx", index_col=0)
df_uncertainty = pd.read_excel("./data/processed/index_uncertainty.xlsx", index_col=0)

df_optimism = df_optimism.select_dtypes(include=['number'])
df_uncertainty = df_uncertainty.select_dtypes(include=['number'])

df_optimism.columns = [f"optimism_{i}" for i in range(len(df_optimism.columns))]
df_uncertainty.columns = [f"uncertainty_{i}" for i in range(len(df_uncertainty.columns))]
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(15, 8))
line_plot = sns.lineplot(data=df_optimism, dashes=False, palette="viridis", linewidth=2.5)

plt.title("Optimism Index for 8 Topics", fontsize=20, weight='bold', pad=20)
plt.xlabel('Date', fontsize=14, labelpad=15)
plt.ylabel('Vol', fontsize=14, labelpad=15)
plt.legend(title="Topics", fontsize='large', title_fontsize='medium')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(15, 8))
line_plot = sns.lineplot(data=df_uncertainty, dashes=False, palette="viridis", linewidth=2.5)

plt.title("Uncertainty Index for 8 Topics", fontsize=20, weight='bold', pad=20)
plt.xlabel('Date', fontsize=14, labelpad=15)
plt.ylabel('Vol', fontsize=14, labelpad=15)
plt.legend(title="Topics", fontsize='large', title_fontsize='medium')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```
